```python
import pandas as pd
import requests 
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
resp = requests.get(url)
resp
data = resp.json()
masterdf= pd.DataFrame(columns=['id','node_id','url','name','color','default','description'])
for i in range(30):
    s = pd.DataFrame(data[i]['labels'], columns=['id','node_id','url','name','color','default','description'])
    masterdf= masterdf.append(s, ignore_index =True)
masterdf
```

```python
# loading data
import pandas as pd
import requests 
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
data = pd.read_json(url)
data = pd.DataFrame(data, columns= ['number','title','user','state','label'])
data.head(5)
```

```python
# split keys and values in columns
import json
from pandas.io.json import json_normalize

user_data = json_normalize(data['user'])
user_data.head(3)
```

```python
# concatination both data set to display all columns
df3 = pd.concat([data[:],user_data[:]],axis = 1)

df3.head(3)
```

```python
df4 = pd.concat([data[['number','title','state','label']],user_data[:]],axis =1)
df4.head(4)
```
